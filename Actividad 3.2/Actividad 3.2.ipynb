```python
import pandas as pd
import numpy as np
```

```python
# Cargar el archivo CSV
df = pd.read_csv('Victoria BC.csv')

# Función para reemplazar valores nulos con la mediana en columnas numéricas
def replace_nulls_with_median(df):
    numcol = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numcol:
        df[col].fillna(df[col].median(), inplace=True)
    return df

# Función para reemplazar valores nulos con el método ffill en columnas de tipo objeto
def replace_nulls_ffill(df):
    objectcols = df.select_dtypes(include=['object']).columns
    for col in objectcols:
        df[col].fillna(method='ffill', inplace=True)
    return df

# Función para reemplazar valores nulos con el método bfill en columnas de tipo objeto
def replace_nulls_bfill(df):
    objectcols = df.select_dtypes(include=['object']).columns
    for col in objectcols:
        df[col].fillna(method='bfill', inplace=True)
    return df

# Función para reemplazar valores nulos con un string concreto en columnas de tipo objeto
def replace_nulls_with_string(df, colname, string):
    df[colname].fillna(string, inplace=True)
    return df

# Función para reemplazar valores nulos con una constante en columnas numéricas
def replace_nulls_with_constant(df, constante):
    numcol = df.select_dtypes(include=['int64', 'float64']).columns
    for col in numcol:
        df[col].fillna(constante, inplace=True)
    return df

# Aplicar las funciones al DataFrame
df = replace_nulls_with_median(df)
df = replace_nulls_ffill(df)
df = replace_nulls_bfill(df)

# Guardar el DataFrame modificado como un nuevo CSV
df.to_csv('Victoria_BC_limpio.csv', index=False)
```
